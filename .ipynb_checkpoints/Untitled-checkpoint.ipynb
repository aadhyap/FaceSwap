```python
import cv2
#import face_recognition
import dlib



#Gets box coordinates of face
def getFace():
    print("started")
    cap= cv2.VideoCapture("face.mp4")
    face_locations = []

    while True:

        ret, frame = cap.read() #frame 
        # Convert the image from BGR color (which OpenCV uses) to RGB   
        # color (which face_recognition uses)
        rgb_frame = frame[:, :, ::-1]
        # Find all the faces in the current frame of video





        frontalFaceDetector = dlib.get_frontal_face_detector()


        # Now the dlip shape_predictor class will take model and with the help of that, it will show 
        


        img= cv2.imread('Capture 8.PNG')
        gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


        allFaces = frontalFaceDetector(gray, 0)

        for face in allFaces:
          x1=face.left()
          y1=face.top()
          x2=face.right()
          y2=face.bottom()
        # Drawing a rectangle around the face
          cv2.rectangle(img, (x1,y1), (x2,y2),(0,255,0),3)
        #cv2_imshow(img)


        faceLandmarkDetector = dlib.shape_predictor(frame)
        landmarks = faceLandmarkDetector(gray, face)
        
        for n in range(0,68):
            x=landmarks.part(n).x
            y=landmarks.part(n).y
            cv2.circle(img, (x, y), 4, (0, 0, 255), -1)
        cv2_imshow(img)

        cv2.waitKey(1000)

'''LOOKING AT A SPECIFIC POINT

x=landmarks.part(31).x
y=landmarks.part(31).y
# Drawing a circle
cv2.circle(img, (x, y), 6, (0, 0, 255), -1)
cv2_imshow(img)


'''

#cv2.imshow('Video', frame)
        
# Wait for Enter key to stop
#if cv2.waitKey(25) == 13:
#break

#def drawPoints():
getFace()
```
